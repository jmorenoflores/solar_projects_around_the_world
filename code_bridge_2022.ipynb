### Fixing the column names in `data`

```python
columns = data.columns.to_list()
converter = lambda x: x.replace(' ', '_')
my_cols = list(map(converter, columns))
my_cols
```

```python
my_cols =list(map(str.lower,my_cols))
data.columns = my_cols
```

```python
#checking data column names
data.info()
```

```python
data = data[[
    'country','project_name','capacity_(mw)','status','start_year', 'retired_year',
     'latitude', 'longitude', 'city', 'state/province', 'region', 'local_area_(taluk,_county)'
     ]]
```
